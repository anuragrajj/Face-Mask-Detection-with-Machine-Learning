```python
capture = cv2.VideoCapture(1)
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_data.detectMultiscale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,255), 4)
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27:
            break

capture.release()
cv2.destroyAllWindows()
        
```
